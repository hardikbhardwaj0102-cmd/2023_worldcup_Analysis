## REPLACING "-" WITH 0 IN DATAFRAME

```python
df1=df1.replace("-",value="0") 
```

## REMOVING *(NOT OUT) FROM HIGH SCORE(HS)

```python
df1['HS'] = df1['HS'].astype(str).str.replace('*','', regex=False)
```

## CONVERING ALL COLUMNS FROM STRING TO NUMERIC

```python
cols = ['Mat','Inns','NO','Runs','HS','Ave','BF','SR','100','50','0','4s','6s']

df1[cols] = df1[cols].apply(pd.to_numeric)
```

# TOP 10 RUN SCORERS

```python
Toprun=df1.sort_values('Runs',ascending=False).head(10)
```

```python
Toprun
```

```python
sns.barplot(data=Toprun,x="Runs",y="Player")
plt.title("TOP 10 RUN SCORERS")
plt.show()
```

# HIGHEST STRIKE RATE OF PLAYERS HAVING MATCHES MORE THAN 5.

```python
topsr=df1[df1["Mat"]>5].sort_values(by="SR",ascending=False).head(10)
```

```python
topsr
```

```python
sns.barplot(data=topsr,x="SR",y="Player")
plt.title("TOP 10 STRIKE RATE")
plt.show()
```

# HIGHEST Average OF PLAYERS HAVING MATCHES MORE THAN 5.

```python
topavg=df1[df1["Mat"]>5].sort_values(by="Ave",ascending=False).head(10)
```

```python
topavg
```

```python
sns.barplot(data=topavg,x="Ave",y="Player")
plt.title("HIGHEST AVERAGE")
plt.show()
```

# CORRELATION BETWEEN RUNS AND STRIKE RATE.

```python
sns.scatterplot(x='Runs', y='SR', data=df1,palette="rainbow",hue="Ave")
plt.title("Runs vs Strike Rate")
plt.show()
```

# REGRESSION PLOT BETWEEN RUNS AND STRIKE RATE.

```python
sns.regplot(data=df1,x="Runs",y="SR")
```

# RUNS DISTRIBUTION

```python
sns.histplot(df1['Runs'], bins=10)
plt.title("Distribution of Runs")
plt.show()
```

# CONCLUSION

#### The analysis identified the top 10 run scorers in the tournament and highlighted players with the highest strike rates and batting averages. Bar plots helped compare player performances, while a histogram showed the distribution of runs among players. Additionally, correlation and regression analysis between runs and strike rate were performed to understand the relationship between scoring speed and total runs.

#### The results suggest that players who maintain a good strike rate often tend to accumulate higher runs, although consistency and match participation also play an important role. Overall, this analysis demonstrates how data visualization and statistical techniques can help uncover meaningful insights from sports performance data.

# Conclusion

#### This analysis explored bowling performance using multiple statistical techniques and visualizations. The results show that bowlers with the highest wickets play a crucial role in match outcomes, while bowlers with lower economy rates are effective in restricting the scoring rate. The analysis of maiden overs further highlights bowlers who maintain consistent line and length, creating pressure on the batting side.

#### Correlation analysis indicates a relationship between economy rate and maiden overs, suggesting that bowlers who bowl tighter overs tend to maintain better economy. However, the correlation between economy and wickets is weaker, implying that wicket-taking ability does not always depend on restricting runs. The regression analysis between runs conceded and wickets suggests that bowlers who bowl more frequently tend to concede more runs but also have greater opportunities to take wickets.

#### Overall, the study shows that bowling effectiveness depends on a balance between run control and wicket-taking ability, and different bowlers contribute in different ways to a team's success.
